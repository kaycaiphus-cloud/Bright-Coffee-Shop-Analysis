```sql
--0. GENERAL QUERY
----------------------------------------------------------------------------
select * from `workspace`.`default`.`bright_coffee_shop_happy` limit 10;

----------------------------------------------------------------------------
--1. CHECKING FIRST AND LAST TRANSACTION DATE
----------------------------------------------------------------------------
select min(transaction_date) First_transaction,
       max(transaction_date) Last_transaction
from `workspace`.`default`.`bright_coffee_shop_happy`;

-----------------------------------------------------------------------------
--2. CHECKING TOTAL NUMBER OF STORES
-----------------------------------------------------------------------------
select distinct store_location
from workspace.default.bright_coffee_shop_happy;

-----------------------------------------------------------------------------
--3. CHECKING SPEND PATTERN
-----------------------------------------------------------------------------
select transaction_qty, unit_price,
        transaction_qty*unit_price as revenue,
case 
      when transaction_qty*unit_price > 5 then 'High'
      when transaction_qty*unit_price > 2 and transaction_qty*unit_price <= 5 then 'Medium'
      else 'Low'
      end as spend_pattern
from `workspace`.`default`.`bright_coffee_shop_happy`;

-----------------------------------------------------------------------------
--4. CHECKING PRODUCT CATEGORIES 
-----------------------------------------------------------------------------
select distinct product_category
from `workspace`.`default`.`bright_coffee_shop_happy`;

-----------------------------------------------------------------------------
--5. CHECKING THE TIMESTAMP
-----------------------------------------------------------------------------
select 
      transaction_id,  
      transaction_date, 
      monthname(transaction_date) as month_of_year,
      dayname(transaction_date) as day_of_week,
      date_format(transaction_time, 'HH:mm:ss') as purchase_time,
      transaction_qty*unit_price as revenue,
      store_location,
      product_category,
      product_type,
      product_detail,
case 
    when dayname(transaction_date) in ('Sat','Sun') then 'Weekend'
    else 'Weekday'
    end as day_type
from `workspace`.`default`.`bright_coffee_shop_happy`;
-----------------------------------------------------------------------------
--6. RUNNING THE BIG/ULTIMATE QUERY
-----------------------------------------------------------------------------
select
      transaction_id,
      store_id,  
      transaction_date, 
      monthname(transaction_date) as month_of_year,
      dayname(transaction_date) as day_of_week,
      date_format(transaction_time, 'HH:mm:ss') as purchase_time,
      transaction_qty*unit_price as revenue,
      store_location,
      product_category,
      product_type,
      product_detail,
---Adding a new column for type of day
case 
    when dayname(transaction_date) in ('Sat','Sun') then 'Weekend'
    else 'Weekday'
    end as day_type,
---Adding new column for time of day
case
      when purchase_time between '00:00:00' and '11:59:59' then '1. Morning'
      when purchase_time between '12:00:00' and '17:59:59' then '2. Afternoon'
      when purchase_time between '18:00:00' and '23:59:59' then '3. Evening'
      end as time_of_day,
---Adding new column for revenue
case
      when transaction_qty*unit_price > 5 then 'High'
      when transaction_qty*unit_price > 2 and transaction_qty*unit_price <= 5 then 'Medium'
      else 'Low'
      end as spend_pattern
from `workspace`.`default`.`bright_coffee_shop_happy`







```
